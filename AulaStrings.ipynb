## O que sao strings em python?

Uma string é uma sequência de caracteres. Em outras palavras, é qualquer conjunto de letras, números, símbolos ou espaços que estejam entre aspas.

Exemplo:



```python
frase = "Python é divertido!"
```

Essa variável frase agora guarda uma string com o texto "Python é divertido!".


# Como criar strings


Você pode criar strings com:
Aspas simples:

nome = 'Ana'


Aspas duplas:
nome = "Ana"


Aspas triplas (para textos de várias linhas):


```python
poema = '''Rosas são vermelhas,
Violetas são azuis.'''
```

Strings são como listas de caracteres
Cada letra (ou caractere) tem uma posição (índice). Por exemplo:
palavra = "Python"


As posições são assim:

Letra.. P Y T H O N

Índice 0 1 2 3 4 5 6

Você pode acessar letras usando colchetes:

print(palavra[0])  # Resultado: 'P'

print(palavra[-1]) # Resultado: 'n' (última letra)

# Semelhanças entre string e lista

Tanto as strings quanto as listas:

✅ Têm índices (começando do 0)

 ✅ Permitem fatiamento (slicing)

 ✅ Podem ser percorridas com loops

 ✅ Possuem comprimento com len()

# Exemplo comparando:


```python
texto = "Python"
lista = ['P', 'y', 't', 'h', 'o', 'n']

print(texto[1])  # Resultado: y
print(lista[1])  # Resultado: y

print(len(texto))  # Resultado: 6
print(len(lista))  # Resultado: 6
```

# Diferença importante: imutabilidade



Strings são imutáveis: você não pode mudar um caractere diretamente.

Listas são mutáveis: você pode alterar qualquer item.

```python
# Isso dá erro!
texto = "Python"
# texto[0] = "J"  ❌ Erro

# Isso funciona
lista = ['P', 'y', 't', 'h', 'o', 'n']
lista[0] = 'J'   # ✅
print(lista)     # ['J', 'y', 't', 'h', 'o', 'n']
```

#Converter entre string e lista
###De string para lista:


```python
texto = "banana"
lista = list(texto)
print(lista)  # ['b', 'a', 'n', 'a', 'n', 'a']
```

##De lista para string:

```python
lista = ['b', 'o', 'l', 'o']
texto = ''.join(lista)
print(texto)  # bolo
```

###Você também pode colocar algo entre os caracteres:

```python
letras = ['a', 'b', 'c']
print('-'.join(letras))  # a-b-c
```

#O que é “len()”:


 len() é uma função embutida  do Python que serve para contar o tamanho de algo — ou seja, quantos elementos tem dentro de uma estrutura como uma string, uma lista, uma tupla, um dicionário, etc.

# Operador + (concatenação)
Ele serve para juntar (concatenar) duas ou mais strings.
## Exemplo:

```python
nome = "Maria"
sobrenome = "Silva"

completo = nome + " " + sobrenome
print(completo)  # Resultado: Maria Silva
```

⚠️ Importante: só dá pra somar strings com strings — não pode misturar com números direto.


```python
idade = 20
# print("Tenho " + idade + " anos")  ❌ Isso dá erro

# Solução:
print("Tenho " + str(idade) + " anos")  ✅
```

# Operador * (repetição)


Ele serve para repetir uma string várias vezes.

 ## Exemplo:


```python
print("ha" * 3)  # Resultado: hahaha
```

## Outro exemplo:


```python
linha = "-" * 30
print(linha)
# Resultado: ------------------------------
```

## Combinação dos dois:


```python
print(("Oi! " + "Tudo bem? ") * 2)
```

# Percorrendo uma string com **for** e **in**

A estrutura básica é:


```python
for letra in minha_string:
    # faça algo com a letra
```

Aqui, a variável letra vai assumir um caractere por vez da string.

## Exemplo simples:


```python
nome = "Ana"

for letra in nome:
    print(letra)
```

## Exibir a posição de cada letra:

```python
palavra = "Python"

for i in range(len(palavra)):
    print(f"Letra {i}: {palavra[i]}")
```

# Principais funções/métodos de string

##1.upper()
 ### Deixa tudo em MAIÚSCULO


```python
nome = "ana"
print(nome.upper())  # ANA
```

#2.lower()
 ### Deixa tudo em minúsculo


```python
texto = "OLÁ"
print(texto.lower())  # olá
```

# 3.capitalize()
 ### Deixa só a primeira letra maiúscula


```python
frase = "python é legal"
print(frase.capitalize())  # Python é legal
```

#4.title()
 ### Deixa a primeira letra de cada palavra maiúscula


```python
titulo = "aprendendo python com strings"
print(titulo.title())  # Aprendendo Python Com Strings
```

#5.strip()
 ### Remove espaços do começo e do fim


```python
mensagem = "  oi  "
print(mensagem.strip())  # oi
```

Tem também:

* .lstrip() → remove só da esquerda
* .rstrip() → remove só da direita


#6.replace(antigo, novo)
 ### Troca partes do texto


```python
frase = "gato é legal"
print(frase.replace("gato", "cachorro"))  # cachorro é legal
```

# 7.find(palavra)
### Retorna o índice onde a palavra começa


```python
texto = "banana"
print(texto.find("na"))  # 2
```

Se não encontrar, retorna -1.


# 8.count(palavra)
### Conta quantas vezes aparece


```python
print("banana".count("a"))  # 3
```

# 9.startswith() e .endswith()
### Verifica se começa ou termina com algo


```python
nome = "ana maria"
print(nome.startswith("ana"))   # True
print(nome.endswith("maria"))   # True
```

#10.split()
### Divide a string em uma lista de palavras


```python
frase = "eu amo python"
palavras = frase.split()
print(palavras)  # ['eu', 'amo', 'python']
```

#11.join()
### Junta elementos de uma lista em uma string só


```python
palavras = ['a', 'b', 'c']
print('-'.join(palavras))  # a-b-c
```

# Sintaxe do slicing


nova_string = string[início:fim:passo]



*   início: onde começa o corte (inclusivo)
*   fim: onde termina o corte (exclusivo)
*   passo: de quantos em quantos você quer pular (opcional)




# Exemplos práticos


```python
texto = "Python"
```

### Do início até um ponto:


```python
print(texto[0:3])  # 'Pyt'
```

### Do meio até o fim:


```python
print(texto[2:])  # 'thon'
```

### Do início até um ponto (sem começo):


```python
print(texto[:4])  # 'Pyth'
```

### String inteira:


```python
print(texto[:])  # 'Python'
```

###  Pulando letras:


```python
print(texto[::2])  # 'Pto' (vai de 2 em 2)
```

### Ao contrário:


```python
print(texto[::-1])  # 'nohtyP' (string invertida)
```
